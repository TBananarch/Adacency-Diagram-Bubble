```python
#create a directed graph
G = nx.DiGraph()
```

```python
# Add nodes and edges based on adjacency relationships
for index, row in df.iterrows():
    program = row['Program']
    area = row['Area']
    adjacency = row['Adjacency']

    G.add_node(program, area=area)
    # Add the node with circular shape if there's no adjacency
    if pd.isna(adjacency):
        G.add_node(program, area=area, shape='o')  # 'o' for circular shape
    else:
        G.add_node(program, area=area, shape='o')
        adjacencies = [adj.strip() for adj in adjacency.split(',')]
        for adj in adjacencies:
            G.add_edge(program, adj)
```
